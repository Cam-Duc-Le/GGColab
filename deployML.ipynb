```python
import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle
from waitress import serve
app = Flask(__name__)
loaded_nn1 = pickle.load(open('/content/nn1.sav', 'rb'))
loaded_nn2 = pickle.load(open('/content/nn2.sav', 'rb'))
loaded_dtree = pickle.load(open('/content/dtree.sav', 'rb'))
loaded_rf = pickle.load(open('/content/rf.sav', 'rb'))
loaded_svm = pickle.load(open('/content/svm.sav', 'rb'))
loaded_logreg = pickle.load(open('/content/logreg.sav', 'rb'))
loaded_nb = pickle.load(open('/content/nb.sav', 'rb'))
@app.route('/')
def home():
    return 

@app.route('/classify',methods=['POST'])
def classify():
    int_features = [int(x) for x in request.form.values()]
    final_features = [np.array(int_features)]
    nn1_arr = loaded_nn1.predict(final_features)        
    nn2_arr = loaded_nn2.predict(final_features)        
    dtree_arr = loaded_dtree.predict(final_features)    
    rf_arr = loaded_rf.predict(final_features)          
    svm_arr = loaded_svm.predict(final_features)        
    logreg_arr = loaded_logreg.predict(final_features)  
    nb_arr = loaded_nb.predict(final_features)   
    tmp_arr = [0,0,0,0]       
    if nn1_arr == 0:
      tmp_arr[0] += 0.9
    elif nn1_arr == 1:
      tmp_arr[1] += 0.9
    elif nn1_arr == 2:
      tmp_arr[2] += 0.9
    elif nn1_arr == 3:
      tmp_arr[3] += 0.931

    if nn2_arr == 0:
      tmp_arr[0] += 0.9
    elif nn2_arr == 1:
      tmp_arr[1] += 0.9
    elif nn2_arr == 2:
      tmp_arr[2] += 0.9
    elif nn2_arr == 3:
      tmp_arr[3] += 0.9

    if dtree_arr == 0:
      tmp_arr[0] += 0.9
    elif dtree_arr == 1:
      tmp_arr[1] += 0.9
    elif dtree_arr == 2:
      tmp_arr[2] += 0.9
    elif dtree_arr == 3:
      tmp_arr[3] += 0.9
  
    if rf_arr == 0:
      tmp_arr[0] += 1
    elif rf_arr == 1:
      tmp_arr[1] += 1
    elif rf_arr == 2:
      tmp_arr[2] += 1
    elif rf_arr == 3:
      tmp_arr[3] += 1
  
    if svm_arr == 0:
      tmp_arr[0] += 0.7
    elif svm_arr == 1:
      tmp_arr[1] += 0.7
    elif svm_arr == 2:
      tmp_arr[2] += 0.7
    elif svm_arr == 3:
      tmp_arr[3] += 0.7
  
    if logreg_arr == 0:
      tmp_arr[0] += 0.7
    elif logreg_arr == 1:
      tmp_arr[1] += 0.7
    elif logreg_arr == 2:
      tmp_arr[2] += 0.7
    elif logreg_arr == 3:
      tmp_arr[3] += 0.7
  
    if nb_arr == 0:
      tmp_arr[0] += 0.6
    elif nb_arr == 1:
      tmp_arr[1] += 0.6
    elif nb_arr == 2:
      tmp_arr[2] += 0.6
    elif nb_arr == 3:
      tmp_arr[3] += 0.6

    if tmp_arr[0]   == max(tmp_arr):
      output = 0
    elif tmp_arr[1] == max(tmp_arr):
      output = 1
    elif tmp_arr[2] == max(tmp_arr):
      output = 2
    elif tmp_arr[3] == max(tmp_arr):
      output = 3
   
    return output


if __name__ == "__main__":
    serve(app, host="0.0.0.0", port=8080)
```
