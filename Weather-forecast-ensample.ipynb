```python
import pandas as pd 
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/Cam-Duc-Le/WeatherMDPfront/master/forecast_data_origin.csv"
data_org = pd.read_csv(url)
print(data_org['condition'].value_counts())
data_org.head()
```

```python
url = "https://raw.githubusercontent.com/Cam-Duc-Le/WeatherMDPfront/master/forecast_data.csv"
data = pd.read_csv(url)
print(data['condition'].value_counts())
data.head()
```

```python
data.dropna()
data.describe()
```

```python
from sklearn import preprocessing
X = np.array(data[['feelslike_c','wind_kph','pressure_mb','humidity','vis_km','cloud']])
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
Y = []
condition = np.array(data['condition'])
count = [0,0,0,0]
for i in range(len(X)):
  if condition[i][10:15] == 'Rainy':
    Y.append(3) 
    count[3]+=1
  elif condition[i][10:15]  == 'Sunny':
    Y.append(2)
    count[2]+=1
  elif condition[i][10:15] == 'Cloud':
    Y.append(1)
    count[1]+=1
  else:
    Y.append(0)
    count[0]+=1

Y = (np.array(Y))
print(X.shape,Y.shape)
count
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,Y_train)
Y_pred = nb.predict(X_test)
print(nb.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, nb.predict(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(multi_class='multinomial',solver='sag',max_iter=2730)
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
print(logreg.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, logreg.predict(X_test)))
```

```python
from sklearn import svm
sm = svm.SVC(kernel='linear')
sm = sm.fit(X_train, Y_train)
Y_pred = sm.predict(X_test)
print(sm.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, sm.predict(X_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion="entropy", max_depth=27)
dtree = dtree.fit(X_train,Y_train)
Y_pred = dtree.predict(X_test)
print(dtree.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, dtree.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=27)
rf = rf.fit(X_train,Y_train)
Y_pred=rf.predict(X_test)
print(rf.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, rf.predict(X_test)))
```

```python
from sklearn.neural_network import MLPClassifier
nn1 = MLPClassifier(hidden_layer_sizes=(16,32,16),activation='relu', solver='adam', max_iter=1000,random_state=1)
nn1.fit(X_train,Y_train)
Y_pred = nn1.predict(X_test)
print(nn1.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, nn1.predict(X_test)))
```

```python
from sklearn.neural_network import MLPClassifier
nn2 = MLPClassifier(hidden_layer_sizes=(27,27,27),activation='relu', solver='adam', max_iter=1000,random_state=1)
nn2.fit(X_train,Y_train)
Y_pred = nn2.predict(X_test)
print(nn2.score(X_train,Y_train))
print(metrics.accuracy_score(Y_pred,Y_test))
print(metrics.classification_report(Y_test, nn2.predict(X_test)))
```

```python
cnf_matrix = metrics.confusion_matrix(Y_test, Y_pred)

class_names=[0,1,2,3] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("bottom")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import accuracy_score
final_output = []
tmp_arr = [0,0,0,0]
nn1_arr = loaded_nn1.predict(X_test)        # *0.931
nn2_arr = loaded_nn2.predict(X_test)        # *0.933
dtree_arr = loaded_dtree.predict(X_test)    # *0.943
rf_arr = loaded_rf.predict(X_test)          # *1.000
svm_arr = loaded_svm.predict(X_test)        # *0.885*0.8
logreg_arr = loaded_logreg.predict(X_test)  # *0.875*0.8
nb_arr = loaded_nb.predict(X_test)          # *0.852*0.8

for i in range(len(nn1_arr)):
  if nn1_arr[i] == 0:
    tmp_arr[0] += 0.9
  elif nn1_arr[i] == 1:
    tmp_arr[1] += 0.8
  elif nn1_arr[i] == 2:
    tmp_arr[2] += 1
  elif nn1_arr[i] == 3:
    tmp_arr[3] += 0.9

  if nn2_arr[i] == 0:
    tmp_arr[0] += 0.9
  elif nn2_arr[i] == 1:
    tmp_arr[1] += 0.8
  elif nn2_arr[i] == 2:
    tmp_arr[2] += 1
  elif nn2_arr[i] == 3:
    tmp_arr[3] += 0.9

  if dtree_arr[i] == 0:
    tmp_arr[0] += 0.9
  elif dtree_arr[i] == 1:
    tmp_arr[1] += 0.8
  elif dtree_arr[i] == 2:
    tmp_arr[2] += 1
  elif dtree_arr[i] == 3:
    tmp_arr[3] += 0.9
  
  if rf_arr[i] == 0:
    tmp_arr[0] += 1
  elif rf_arr[i] == 1:
    tmp_arr[1] += 0.8
  elif rf_arr[i] == 2:
    tmp_arr[2] += 1.2
  elif rf_arr[i] == 3:
    tmp_arr[3] += 1
  
  if svm_arr[i] == 0:
    tmp_arr[0] += 0.7
  elif svm_arr[i] == 1:
    tmp_arr[1] += 0.5
  elif svm_arr[i] == 2:
    tmp_arr[2] += 0.8
  elif svm_arr[i] == 3:
    tmp_arr[3] += 0.7
  
  if logreg_arr[i] == 0:
    tmp_arr[0] += 0.7
  elif logreg_arr[i] == 1:
    tmp_arr[1] += 0.5
  elif logreg_arr[i] == 2:
    tmp_arr[2] += 0.8
  elif logreg_arr[i] == 3:
    tmp_arr[3] += 0.7
  
  if nb_arr[i] == 0:
    tmp_arr[0] += 0.6
  elif nb_arr[i] == 1:
    tmp_arr[1] += 0.5
  elif nb_arr[i] == 2:
    tmp_arr[2] += 0.7
  elif nb_arr[i] == 3:
    tmp_arr[3] += 0.6

  if tmp_arr[0]== max(tmp_arr):
    final_output.append(0)
  elif tmp_arr[1]== max(tmp_arr):
    final_output.append(1)
  elif tmp_arr[2]== max(tmp_arr):
    final_output.append(2)
  elif tmp_arr[3]== max(tmp_arr):
    final_output.append(3)

  tmp_arr = [0,0,0,0]

print(len(final_output),max(final_output))
print(accuracy_score(final_output, Y_test))
```
